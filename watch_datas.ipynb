```python
import numpy as np
import scipy.io as sio
import matplotlib as mpl
import matplotlib.pyplot as plt
import torch
```

```python

dataset = "urban1"  # 选择数据集
data_path = "./data/" + dataset + "_dataset.mat"  # 构建数据路径
data = sio.loadmat(data_path)  # 加载 .mat 文件数据
print(data.keys())  # 查看数据的键
Y = torch.from_numpy(data['Y'].T)
A = torch.from_numpy(data['A'].T)
M = torch.from_numpy(data['M'])
M1 = torch.from_numpy(data['M1'])
Y.shape, A.shape, M.shape, M1.shape
# print(data)
```

```python
# 创建图形和子图
fig, ax = plt.subplots()

# 使用 jet 颜色映射显示数据
cax = ax.imshow(Y[:,100].reshape(307,307), cmap='jet')

# 添加颜色条
fig.colorbar(cax)

# 显示图形
plt.show()
```
